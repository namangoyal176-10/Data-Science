```python
class Complex:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        a = self.x + other.x
        b = self.y + other.y
        return Complex(a, b)

    def display(self):
        print(self.x, "+ i", self.y)


c1 = Complex(2, 3)
c2 = Complex(4, 5)

c3 = c1 + c2

c3.display()
```
