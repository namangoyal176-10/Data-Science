```python
import numpy as np
n = int(input("Number of items : "))

prices = []

for i in range(n):
    prices.append(int(input()))

prices = np.array(prices)

print("Prices =", prices)
print("")
print("Total Bill    =" , np.sum(prices))
print("Average Price =" , np.mean(prices))
print("Std Dev       =" , np.std(prices))
print("Unique Prices =" , np.unique(prices))
print("Sorted Prices =" , np.sort(prices))
print("Minimum Price =" , np.min(prices))
print("Maximum Price =" , np.max(prices))


print("Items > 100   =" , prices[np.where(prices > 100)])
```
